```
# Q19)Read the input as Number and print the output "Yes" or "No".
# Test cases
#  145 -- Yes
#  123 -- No
# Hint:Individual digit factorial sum is same as original number then "Yes" otherwise "No".
```

```
# Answer code is

#include<stdio.h>
int factorial(int x)
{
	int i,f=1;
	for(i=x;i>0;i--)
	{
		f=f*i;
	}
	return f;
}
int main()
{
	int a,r,sum=0,temp;
	scanf("%d",&a);
	temp=a;
	while(a!=0)
	{
		r=a%10;
		factorial(r);
		sum=sum+factorial(r);
		a=a/10;		
	}
	if(temp == sum){
		printf("Yes");
	}
	else{
		printf("No");
	}
}
```

```
# Q21)Read the inut as number and print the output as sum of the (digits to power the count digits.)
# Test cases
#  121 -- 10
#  1221 -- 34
# Hint:121 number has total 3 digits so 1^3+2^3+1^3
```

```
# Answer code is

#include<stdio.h>
#include<math.h>
int digitCount(int x)
{
	int count=0;
	while(x!=0)
	{
		count++;
		x=x/10;		
	}
	return count;	
}
int main()
{
	int a,count,sum=0,r;
	scanf("%d",&a);
	count=digitCount(a);
	while(a!=0)
	{
		r=a%10;
		sum=sum+pow(r,count);
		a=a/10;
	}
	printf("%d",sum);
	return 0;
}
```
