### Binary search

```
# Binary Search code

    #include<stdio.h>
int binarySearch(int a[],int n,int tar)
{
	int mid,leftIndex=0,rightIndex=n-1;
	while(leftIndex<=rightIndex){
		mid=leftIndex+(rightIndex-leftIndex)/2;
		if(a[mid]==tar){
			return mid;
		}
		if(a[mid]<tar){
			leftIndex=mid+1;
		}
		else{
			rightIndex=mid-1;
		}
	}
	return -1;
}
int main()
{
	int resIndex;
	int x[9]={1,11,18,25,45,76,89,99,122};
	resIndex=binarySearch(x,9,122);
	if(resIndex>=0){
		printf("Item Found");
	}
	else{
		printf("Item not Found");
	}
	return 0;
}
```

### Bubble Sort
- It compares the adjacent items,if first item is greater than second item then swap those both items
- complexity is O(n^2)

```
# Bubble Sort code

    #include<stdio.h>
void bubbleSort(int a[],int n)
{
	int i,j,temp;
	for(i=0;i<n-1;i++){
		for(j=0;j<n-1;j++){
			if(a[j]>a[j+1]){
				temp =a[j];
				a[j]=a[j+1];
				a[j+1]=temp;
			}
		}
	}
	for(i=0;i<n;i++){
		printf("%d ",a[i]);
	}
}
int main()
{
	int x[6] ={19,2,25,18,12,5};
	bubbleSort(x,6);
	return 0;
}
```

### Selection Sort
- Selection sort will divide the given list into two sub parts
   - sorted array
   - unsorted array
- This algorithm works repeatedly by finding the least element of the list and placing at first positions 
- colmplexity is O(n^2)

```
# Selection Sort code

    #include<stdio.h>
void selectionSort(int a[],int n)
{
	int i,j,k,temp;
	for(i=0;i<n-1;i++){
		k=i;
		for(j=i;j<n;j++){
			if(a[j]<a[k]){
				k = j;//index interchange
			}
		}
		//swap the i and k
		temp=a[i];
		a[i]=a[k];
		a[k]=temp;
	}
	for(i=0;i<n;i++){
		printf("%d ",a[i]);
	}
}
int main()
{
	int x[100],i,n;
	scanf("%d",&n);
	for(i=0;i<n;i++){
		scanf("%d",&x[i]);
	}
	selectionSort(x,n);
	return 0;
}
```

### Insertion Sort
- complexity is O(n^2)
- This algorith works similar to playing caed game algorith
- it used to sort few items of the data list

```
# Insertion Sort code

    #include<stdio.h>
void insertionSort(int a[],int n)
{
	int i,j,k,temp;
	for(i=1;i<n;i++){
		k=a[i];
		j=i-1;
		while(j>=0 && a[j]>k){
			a[j+1]=a[j];
			j=j-1;
		}
		a[j+1]=k;
	}
		for(i=0;i<n;i++){
			printf("%d ",a[i]);
		}
	}
int main()
{
	int x[100],i,n;
	scanf("%d",&n);
	for(i=0;i<n;i++){
		scanf("%d",&x[i]);
	}
	insertionSort(x,n);
	return 0;
}
```

# Stacks
- it uses the data type array
- Stack is a memory storage
- Stack is responsible for static memory
- It looks like one sided open box
- push-inserts element
- pop-removes element
- peek-it is the top most element in stack

```
# Stack Operations Code

    #include<stdio.h>
#include<stdlib.h>
int main()
{
int a[50],top=-1,ch,n,i;
printf("enter the size");
scanf("%d",&n);

while(1){
printf("-----STACK------\n1.PUSH\n2.POP\n3.PEEK\n4.DISPLAY\n5.EXIT\n");
printf("enter your choice\n");
scanf("%d",&ch);
switch(ch){
case 1:
if(top<n){
top++;
printf("\nEnter item value");
scanf("%d",&a[top]);
printf("\nitem pushed");
}
else{
printf("\nstack is full");
}
break;
case 2:
if(top!=-1){
a[top]=0;
top--;
printf("\nitem poped");
}
else{
printf("\nstack is empty");
}
break;
case 3:
if(top!=-1){
printf("\nPeek item of stack");
printf("\n%d",a[top]);
}
else{
printf("\nstack is empty");
}
break;
case 4:
if(top!=-1){
for(i=top;i>=0;i--){
printf("%d ",a[i]);
}
}
else{
printf("\nstack is empty");
}
break;
case 5:
printf("EXIT");
return 0;

}

}
}
```

```
# LinkedList operations code

    #include<stdio.h>
#include<stdlib.h>
#include<conio.h>
struct node
{
	int data;
	struct node *next;
}*s,*l,*c,*p;
void insertatbegin();
void insertatposition();
void insertatend();
void deletestartnode();
void deleteatend();
void deletebyvalue();
void displaynode();
void searchnode();
void lengthofl();
int main()
{
	int x,y,z;
	s=l=c=p=NULL;
while(1)
{
	printf("\n1.INSERT\n2.REMOVE\n3.DISPLAY\n4.SEARCHELEMENT\n5.LENGTHOFLINKEDLIST\n6.EXIT\n ENTER YOUR CHOICE :\n");
	scanf("%d",&x);
switch(x)
{
	case 1:printf("\n1.INSERT AT BEGINNING\n2.INSERT AT POSITION\n3.INSERT AT END\n ENTER YOUR CHOICE :\n");
	scanf("%d",&y);
		switch(y)
		{
			case 1:insertatbegin();
			break;
			case 2:insertatposition();
			break;
			case 3:insertatend();
			break;
default:printf("\n---INVALID INPUT---\n");
}
break;
case 2:printf("\n1.DELETE AT BEGINNING\n2.DELETE A VALUE\n3.DELETE AT END\n ENTER YOUR CHOICE :\n");
	scanf("%d",&z);
		switch(z)
		{
			case 1:deletestartnode();
			break;
			case 2:deletebyvalue();
			break;
			case 3:deleteatend();
			break;
			default:printf("\n---INVALID INPUT---\n");
		}
break;
case 3:displaynode();
break;
case 6:printf("\nPROGRAM IS TERMINATED\n");
return 0;
break;
case 4:searchnode();
break;
case 5:lengthofl();
break;
default:printf("\n---INVALID INPUT---\n");
}
}
return 0;
}
void insertatbegin()
{
	struct node *nn=(struct node *)malloc(sizeof(struct node));
	printf("\nENTER THE VALUE\n");
	scanf("%d",&nn->data);
	nn->next=NULL;
	if(s==NULL)
		{
		s=l=nn;
		}
	else
		{
		nn->next=s;
		s=nn;
		}
}
void insertatend()
{
	struct node *nn=(struct node *)malloc(sizeof(struct node));
	printf("\nENTER THE VALUE\n");
	scanf("%d",&nn->data);
	nn->next=NULL;
	if(s==NULL)
	{
	s=l=nn;
	}
	else
		{
		l->next=nn;
		l=nn;
		}
}
void insertatposition()
{
	int po,i=0;
	struct node *nn=(struct node *)malloc(sizeof(struct node));
	printf("\nENTER THE VALUE\n");
	scanf("%d",&nn->data);
	nn->next=NULL;
	if(s==NULL){
	s=l=nn;		
	}
	else{
	printf("\nENTER THE POSITION\n");
	scanf("%d",&po);
	if(po==1)
	{
	nn->next=s;
	s=nn;
	}
	else
	{
	for(c=s;i<po;p=c,c=c->next,i++);
	nn->next=c;
	p->next=nn;
	}
	}	
}
void  deletestartnode()
{
	if(s==NULL)
	{
	printf("\nTHE LIST IS EMPTY\n");
	}
	else
		{
		if(s->next==NULL)
			{
			s=l=NULL;
			printf("\nELEMENT IS DELETED\n");
			}
		else
		{
		c=s;
		s=s->next;
		c->next=NULL;
		c=NULL;
		printf("\nELEMENT IS DELETED FROM START\n");
		}
		}
}
void  deletebyvalue()
{
	int si;
	printf("\nENTER A VALUE\n");
	scanf("%d",&si);
	if(s->next==NULL&&si==s->data)
		{
		s=l=NULL;
		printf("\nNODE IS DELETED\n");
		}
	else if(si==s->data)
		{
		c=s;
		s=s->next;
		c->next=NULL;
		printf("\nNODE IS DELETED\n");
		}
	else if(si==l->data)
		{
		for(c=s;c!=l;p=c,c=c->next);
		l=p;
		p->next=NULL;
		c=NULL;
		p=NULL;
		printf("\n NODE IS DELETED\n");
		}
	else
		{
		for(c=s;c!=NULL&&c->data!=si;p=c,c=c->next)
		{
	if(c==NULL)
		{
		printf("\nNODE IS NOT FOUND\n");
		}
	else
		{
		p->next=c->next;
		c->next=NULL;
		c=NULL;
		printf("\nNODE IS DELETED\n");
		}
		}
		}
}
void  deleteatend()
{
	if(s==NULL)
		{
		printf("\nTHE LIST IS EMPTY\n");
		}
	else
		{
	if(s->next==NULL)
		{
		s=l=NULL;
		printf("\nELEMENT IS DELETED\n");
		}
	else
		{
		for(c=s;c!=l;p=c,c=c->next);
		l=p;
		p->next=NULL;
		p=NULL;
		c=NULL;
		printf("\nELEMENT IS DELETED FROM LAST\n");
		}
		}
}
void displaynode()
{
	if(s==NULL)
		{
		printf("\nTHE LIST IS EMPTY\n");
		}
	else
	{
	printf("\n-------------------\n");
	printf("\n----LINKED LIST----\n");
	for(c=s;c!=NULL;c=c->next)
	{
	printf("%d ",c->data);
	}
	}
}
void searchnode()
{
	int se,pos=1,i=1;
	printf("enter element to be searched:\n");
	scanf("%d",&se);
	for(c=s;c!=NULL;c=c->next)
	{
	if(c->data!=se)
		{
		pos++;
		i++;
		}
	else
		{
		printf("\nELEMENT IS PRESENT AT:%d \n",pos);
		pos++;
		i--;
		}
	}
	if(pos==i)
	{
	printf("\nELEMENT IS NOT PRESENT\n");
	}
}	
void lengthofl()
{
	int le=0;
	for(c=s;c!=NULL;c=c->next)
	{
		le++;
	}
	printf("\nLENGTH IS %d\n",le);
}
```
