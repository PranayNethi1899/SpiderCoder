## for loop
syntax: for x in range(parameters):
statements

```
def printseries(lb,ub):
    for x in range(lb,ub+1):
        print(x,end=" ")
    return
printseries(1,10)        
```

```
#function to print alternate numbers
#[500,520]--500 502 504 506.....520
def printeven(lb,ub,):
    for x in range(lb,ub+1):
        if(x%2==0):
            print(x,end=" " )
    return
printeven(500,520)        
```

```
#same program with the help of jump statement
for x in range(500,520+1,2):
    print(x,end=" ")    
```

```
contacts = {} #this is empty dictionary
def addcontact(name,phone):
    if (name not in contacts):
        contacts[name]= phone
        print("contact detils are added sucessfully")
    else:
        print("name alredy exists")          
    return
addcontact("kishore",'9553093619')    
```
